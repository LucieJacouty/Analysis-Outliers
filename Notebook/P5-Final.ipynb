# Nettoyage données web (suppressions doublons, lignes nulles, colonnes inutiles)

```python
def clean(df):
    subset = ['sku']
    labels = ['virtual', 'downloadable', 'rating_count', 'average_rating', 'tax_status', 'tax_class',
        'post_content', 'post_status', 'comment_status', 'ping_status', 'post_password', 'post_content_filtered',
        'post_parent', 'guid', 'menu_order', 'post_mime_type', 'comment_count']
    df.drop_duplicates(subset, inplace=True)
    df.drop(columns=labels, inplace=True)
    df.dropna(0,'all', inplace=True)
    return df
```

```python
export = web.copy()
export = clean(export)

print("unique: \n",export.nunique(axis=0))
print("\ncount: \n",export.count())
print("\nnull: \n",export.isna().sum())
```

# Jointure des 3 fichiers

```python
merge = erp.merge(liaison,on='product_id').merge(export,left_on='id_web', right_on='sku')

data = merge.copy()
print("unique: \n",data.nunique(axis=0))
print("\ncount: \n",data.count())
print("\nnull: \n",data.isna().sum())
data
```

```python
data['CA'] = data['price']*data['total_sales']

print("Le chiffre d'affaire total est de",data['CA'].sum(),"€")
data['CA'].head(30)
```

Courbe de Lorenz et indice de gini

```python
prices = data['price'].values
n = len(prices)
lorenz = np.cumsum(np.sort(prices)) / prices.sum()
lorenz = np.append([0],lorenz) 

plt.axes().axis('equal')
xaxis = np.linspace(0-1/n,1+1/n,n+1)
plt.plot(xaxis,lorenz,drawstyle='steps-post')
plt.show()
```

```python
AUC = (lorenz.sum() -lorenz[-1]/2 -lorenz[0]/2)/n 
S = 0.5 - AUC
gini = 2*S
print("L'indice de gini est de :\n", gini)
```

## Recherche des outliers :

### Méthode z score

```python
data['z_score'] = (data['price']-data['price'].mean())/data['price'].std(ddof=0)
data_out = data.loc[data['z_score'] > 2]

print("Outliers :\n")
data_out['price'].hist()
plt.show()
data_out.boxplot(column="price")
plt.show
```

```python
print('Le nombre de valeurs ayant un z score supérieur à 2 est de', data_out.shape[0])
```

```python
plt.figure(figsize=(15,10))
sns.barplot(data=data_out.sort_values(by=['price']), x='price', y='post_title', palette='Blues')
plt.xlabel('Prix', fontsize=15)
plt.ylabel('Nom', fontsize=15)
plt.title('Nom et prix des vins les plus chers (z score)', fontsize=20)
plt.tight_layout()
```

### Méthode IQ

```python
Q1 = np.percentile(data['price'], 25)
Q3 = np.percentile(data['price'], 75)
IQ = Q3-Q1
low_lim = Q1-1.5*IQ
upp_lim = Q3+1.5*IQ

print("La limite inférieure est de", round(low_lim))
print("La limite supérieure est de", round(upp_lim))
```

```python
prices_low = data[data['price']<=low_lim]
prices_upp = data[data['price']>=upp_lim]

print('Le nombre de valeurs sortant de la limite inférieure est de', prices_low.shape[0])
print('Le nombre de valeurs sortant de la limite supérieure est de', prices_upp.shape[0])
```

```python
print("Outliers: \n")
prices_upp['price'].hist()
plt.show()
prices_upp.boxplot(column="price")
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.barplot(data=prices_upp.sort_values(by=['price']), x='price', y='post_title', palette='Blues')
plt.xlabel('Prix', fontsize=15)
plt.ylabel('Nom', fontsize=15)
plt.title('Nom et prix des vins les plus chers (IQ)', fontsize=20)
plt.tight_layout()
```
